```python

from time import sleep
import os
#listsandstuff
pizza = {'Margherita': {'Regular': 109, 'Medium': 239, 'Large': 449}, 
         'Farmhouse': {'Regular': 259, 'Medium': 459, 'Large': 689}, 
         'Cheese n Corn': {'Regular': 209, 'Medium': 379, 'Large': 609}, 
         'Tandoori Paneer': {'Regular': 299, 'Medium': 549, 'Large': 799}, 
         'Cheese Dominator': {'Regular': 349, 'Medium': 639, 'Large': 959}}

topping = {'ONION': 60, 'BLACK OLIVE': 60, 'TOMOTO': 60}; crust = {'CHEESE BURST': 50, 'FRESH PAN': 80, 'WHEAT': 110}

mypizza = []; mypizzasize = []; mytopping = []; mycrust = []; subtotal = []
price = 0; nooforder = 0
#listandstuff   

welcome_message = "WELCOME TO PIZZA ORDERING WEBSITE"
for i in range(33):
    print(welcome_message[i], sep=' ', end=' ', flush=True); sleep(0.1)
print("\nName: Sanket Patil\nRoll no: 135\tDiv: B\nPR ASSIGNMENT 2\n")
sleep(1)

ask = input("\nWould you like to order?(y/n)?: ")
if ask == 'y':
    order = True
    name = input("Please enter your name: ")
    address = input("Please Enter your location: ")
    pincode = int(input("Enter your pin code: "))
    print("Thankyou...")
    sleep(1)
    clear = lambda: os.system('cls')
    clear()
else:
    order = False
    print("Thankyou...")

while order == True:
    nooforder += 1
    if nooforder == 1:
        menu_message = "  PIZZAS\t\t\tSIZES\t\t\t\tTOPPINGS\t\t\t\tCRUST"
        for i in range(37):
            print(menu_message[i], sep='', end=' ', flush=True); sleep(0.1)
    else:
        print("  \tPIZZAS\t\t\t\tSIZES\t\t\t\t\tTOPPINGS\t\t\t\tCRUST")

    print('''
---------------------- ------------------------------------------- -------------------------------------------- -------------------------------------------------
 1. Margherita        | REGULAR(109₹), MEDIUM(239₹), LARGE(449₹)  | ONION(60₹), BLACK OLIVE(60₹), TOMOTO(60₹)  | CHEESE BURST(50₹), FRESH PAN(80₹), WHEAT(110₹) |
                      |                                           |                                            |                                                |
 2. Farmhouse         | REGULAR(259₹), MEDIUM(459₹), LARGE(689₹)  | ONION(60₹), BLACK OLIVE(60₹), TOMOTO(60₹)  | CHEESE BURST(50₹), FRESH PAN(80₹), WHEAT(110₹) |
                      |                                           |                                            |                                                |
 3. Cheese n Corn     | REGULAR(209₹), MEDIUM(379₹), LARGE(609₹)  | ONION(60₹), BLACK OLIVE(60₹), TOMOTO(60₹)  | CHEESE BURST(50₹), FRESH PAN(80₹), WHEAT(110₹) |
                      |                                           |                                            |                                                |
 4. Tandoori Paneer   | REGULAR(299₹), MEDIUM(549₹), LARGE(799₹)  | ONION(60₹), BLACK OLIVE(60₹), TOMOTO(60₹)  | CHEESE BURST(50₹), FRESH PAN(80₹), WHEAT(110₹) |
                      |                                           |                                            |                                                |
 5. Cheese Dominator  | REGULAR(349₹), MEDIUM(639₹), LARGE(959₹)  | ONION(60₹), BLACK OLIVE(60₹), TOMOTO(60₹)  | CHEESE BURST(50₹), FRESH PAN(80₹), WHEAT(110₹) |
---------------------- ------------------------------------------- -------------------------------------------- -------------------------------------------------
(NOTE: CGST: 2.5% and SGST: 2.5% is applicable)
        WELCOME..
    ''')
    pizza_choice = int(input("\tSelect your pizza(1, 2, 3...): "))  
    pizzachoice = list(pizza.keys())[pizza_choice - 1]
    mypizza.append(pizzachoice)
    print("\nGreat Choice!")
    
    print("\n\t1. REGULAR, 2. MEDIUM, 3. LARGE")
    pizza_size = int(input("\tSelect your pizza size: "))
    pizzasizea = list(pizza.values())[pizza_choice - 1]
    pizzasize = list(pizzasizea.keys())[pizza_size - 1]
    mypizzasize.append(pizzasize)
    price = list(pizzasizea.values())[pizza_size - 1]
    
    get_topping = input("Looks good, Would you like to add a topping?(y/n): ")
    if get_topping == "y": 
        print('''
            1. ONION, 2. BLACK OLIVE, 3. TOMOTO
            (NOTE: WE PROVIDE ONLY ONE TOPPING PER PIZZA!)
            ''')
        pizza_topping = int(input("Select your topping: "))
        pizzatopping = list(topping.keys())[pizza_topping - 1]
        pizzatoppingv = list(topping.values())[pizza_topping - 1]
        mytopping.append(pizzatopping)
        price += pizzatoppingv

    else:
        mytopping.append("none")
    get_crust = input("Cool, Would you like to customize your pizza crust?(y/n): ")
    if get_crust == "y":
        print('''
            1. CHEESE BURST, 2. FRESH PAN PIZZA, 3. WHEAT''')
        pizza_crust = int(input("would you like to add more pizzas?(y/n): "))
        pizzacrust = list(crust.keys())[pizza_crust - 1]
        pizzacrustv = list(crust.values())[pizza_crust - 1]
        mycrust.append(pizzacrust)
        price += pizzacrustv
        

    else:
        mycrust.append("none")
    subtotal.append(price)
    print("\n Great!")

    more_pizzas = input("would you like to add more pizzas?(y/n): ")
    clear = lambda: os.system('cls')
    clear()
    if more_pizzas == 'y':
        order = True
    else:
        order = False
#GST
original_price = sum(subtotal)
gst = 5*original_price/100
net_price = original_price + gst
#print order
print("##################\n")
print("Name: ", name)
print("Address: ", address)
print("Pincode: ", pincode)
print("Your Order:\n")
for i in range(nooforder):
    print("#",(i+1))
    print("Pizza: ", mypizza[i])
    print("Size: ", mypizzasize[i])
    print("Topping: ", mytopping[i])
    print("Crust: ", mycrust[i])
    print("Price: ", subtotal[i])
    print()
print("----------------------")
print("Price: ", net_price,"₹")
print("----------------------")
print("Thankyou for ordering from us...")
print("\n##################\n")
```
